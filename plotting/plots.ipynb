```python
from matplotlib import pyplot as plt
import pandas as pd
results = pd.read_csv("optimal_parameter/Tiled_Multiplication_optimal_parameter.csv")
results
```

```python
results_1024 = results.iloc[26:]
ns = results_1024["n"].values
param = results_1024["param"].values
means = results_1024["Mean"].values
stdev = results_1024["Standard Deviation"].values
```

```python
results_256 = results.loc[results["n"]==256]
results_512 = results.loc[results["n"]==512]
results_1024 = results.loc[results["n"]==1024]
results_2048 = pd.read_csv("optimal_parameter/Tiled_Multiplication_optimal_parameter_large.csv")
param_2 = results.loc[results["n"]==1024]
```

```python
fig = plt.figure()
ax = fig.gca()
ax.errorbar(results_256["param"].values, results_256["Mean"].values,results_256["Standard Deviation"].values, capsize = 3.0, marker = 'o')
ax.errorbar(results_512["param"].values, results_512["Mean"].values,results_512["Standard Deviation"].values, capsize = 3.0, marker = 'o')
ax.errorbar(param, means, stdev, capsize = 3.0, marker = 'o')
ax.errorbar(results_2048["param"].values, results_2048["Mean"].values,results_2048["Standard Deviation"].values, capsize = 3.0, marker = 'o')

ax.set_title("Average runtime for different values of s")
ax.set_xlabel('s')
ax.set_ylabel('Time (seconds)')
ax.set_yscale('log', base=10)
ax.set_xscale("log", base=2)
ax.legend(['n = 256', 'n = 512', 'n = 1024', 'n = 2048'])
plt.savefig('tiled_plot.pdf')
```
